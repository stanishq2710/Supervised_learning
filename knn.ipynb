```python
# knn(K nearest neighbour) model ---> here the outcome is predicted by the outcome values of it s neighbour
import pandas as pd
from sklearn.datasets import load_iris
from  sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
x = pd.DataFrame(iris.data, columns =iris.feature_names)
y = pd.Series(iris.target)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred),annot=True, cmap= "Greens")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - KNN")
plt.show()
# here the accuracy comes as 100% that means the model is overfitted
```
